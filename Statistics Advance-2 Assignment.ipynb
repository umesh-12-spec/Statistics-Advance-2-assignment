## Theorical Question and Answers

1.  What is hypothesis testing in statistics?
   - Hypothesis testing is a method in statistics used to make decisions about a population based on sample data. It starts with two hypotheses: the null hypothesis (Hâ‚€), which suggests no effect or difference, and the alternative hypothesis (Hâ‚), which suggests there is an effect or difference.

     A significance level (Î±) is chosen, usually 0.05, which indicates the probability of rejecting the null hypothesis when it is true. A test statistic is then selected based on the data type and hypothesis, such as a t-test or z-test.

     Using the sample data, the test statistic and p-value are calculated. The p-value shows the probability of obtaining the observed results if the null hypothesis is true. If the p-value is less than the significance level, the null hypothesis is rejected in favor of the alternative hypothesis. Otherwise, the null hypothesis is not rejected.

     Hence, hypothesis testing helps determine if there is enough evidence to support a claim about a population based on sample data.

2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?
   - The null hypothesis (Hâ‚€) is a statement in statistics that assumes there is no effect or no difference in a population. It serves as the default or baseline assumption that researchers aim to test. For example, if you are testing a new drug, the null hypothesis might state that the drug has no effect on patients.

     On the other hand, the alternative hypothesis (Hâ‚) is a statement that contradicts the null hypothesis. It suggests that there is an effect or a difference. In the drug example, the alternative hypothesis would state that the drug does have an effect on patients.

       In hypothesis testing, researchers collect sample data and use statistical methods to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis. If the evidence is strong enough, the null hypothesis is rejected, indicating that the alternative hypothesis is more likely to be true. If the evidence is not strong enough, the null hypothesis is not rejected, meaning there is not enough evidence to support the alternative hypothesis.



3. What is the significance level in hypothesis testing, and why is it important?
   - The significance level (Î±) in hypothesis testing is a threshold used to determine whether to reject the null hypothesis. It represents the probability of making a Type I error, which is rejecting the null hypothesis when it is actually true. Common significance levels are 0.05, 0.01, and 0.10.

     The significance level is important because it helps control the risk of making incorrect conclusions. By setting a significance level, researchers can decide how strong the evidence must be before rejecting the null hypothesis. A lower significance level means stricter criteria for rejecting the null hypothesis, reducing the risk of a Type I error but increasing the risk of a Type II error (failing to reject a false null hypothesis).

     Hence, the significance level is a crucial part of hypothesis testing as it balances the risks of making errors and ensures that conclusions are based on strong evidence.



4. What does a P-value represent in hypothesis testing?
   - In hypothesis testing, the p-value represents the probability of obtaining the observed results, or more extreme results, if the null hypothesis is true. It helps determine the strength of the evidence against the null hypothesis.

     A low p-value (typically less than the chosen significance level, such as 0.05) indicates that the observed data is unlikely under the null hypothesis, leading to the rejection of the null hypothesis in favor of the alternative hypothesis. Conversely, a high p-value suggests that the observed data is consistent with the null hypothesis, and there is not enough evidence to reject it.

     Hence, the p-value quantifies the evidence against the null hypothesis, guiding researchers in making decisions based on their data.



5.  How do you interpret the P-value in hypothesis testing?
   - Interpreting the p-value in hypothesis testing involves understanding its role in evaluating the strength of evidence against the null hypothesis. Here's how to interpret it:
  
     Low P-value (typically â‰¤ 0.05): This indicates strong evidence against the null hypothesis. It suggests that the observed data is unlikely to have occurred by random chance if the null hypothesis were true. Therefore, you reject the null hypothesis in favor of the alternative hypothesis (Hâ‚).

     High P-value (> 0.05): This indicates weak evidence against the null hypothesis. It suggests that the observed data is consistent with the null hypothesis, and there is not enough evidence to reject it. Therefore, you fail to reject the null hypothesis.

     Hence, the p-value helps determine whether the observed data provides sufficient evidence to reject the null hypothesis and support the alternative hypothesis.



6. What are Type 1 and Type 2 errors in hypothesis testing?
   - In hypothesis testing, Type I and Type II errors are two types of mistakes that can occur when making decisions based on sample data:

    Type I Error (False Positive):

    This occurs when the null hypothesis (Hâ‚€) is rejected when it is actually true. In other words, it is the incorrect conclusion that there is an effect or difference when there isn't one. The probability of making a Type I error is denoted by the significance level (Î±), typically set at 0.05.

     Type II Error (False Negative):

    This occurs when the null hypothesis (Hâ‚€) is not rejected when it is actually false. In other words, it is the incorrect conclusion that there is no effect or difference when there actually is one. The probability of making a Type II error is denoted by Î² (beta).

    Balancing these errors is crucial in hypothesis testing. Reducing the significance level (Î±) decreases the likelihood of a Type I error but increases the likelihood of a Type II error, and vice versa. Researchers aim to minimize both errors to make accurate and reliable conclusions.



7.  What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
   - In hypothesis testing, the difference between a one-tailed and a two-tailed test lies in the direction of the test and the hypotheses being tested:

     One-Tailed Test:

     A one-tailed test examines whether the sample data is significantly greater than or less than a specified value. It tests for an effect in only one direction.

     The alternative hypothesis (Hâ‚) is directional, indicating that the parameter is either greater than or less than the null hypothesis value.

     Two-Tailed Test:

     A two-tailed test examines whether the sample data is significantly different from a specified value, regardless of the direction. It tests for an effect in both directions.

     The alternative hypothesis (Hâ‚) is non-directional, indicating that the parameter is different from the null hypothesis value.

     Hence, a one-tailed test looks for an effect in one direction, while a two-tailed test looks for an effect in both directions. The choice between the two depends on the research question and the hypotheses being tested.

8.  What is the Z-test, and when is it used in hypothesis testing?
   - The Z-test is a statistical method used to determine if there is a significant difference between sample data and a population parameter or between two sample means. It relies on the standard normal distribution (Z-distribution) and is typically used when the sample size is large (n > 30) and the population variance is known.

      There are three main types of Z-tests:

      Two-Sample Z-test: This test compares the means of two independent samples. For instance, it can be used to test if the average test scores of students from two different schools are significantly different.

      Proportion Z-test: This test compares the sample proportion to a known population proportion or compares the proportions of two independent samples. For example, it can be used to test if the proportion of voters supporting a candidate in a sample is different from the known proportion in the population.

      The Z-test is useful for hypothesis testing when the sample size is large, and the data follows a normal distribution. It helps determine if the observed differences are statistically significant or if they could have occurred by random chance.

12. What is the T-test, and how is it used in hypothesis testing?
   - The T-test is a statistical method used to determine if there is a significant difference between the means of two groups. It's useful when dealing with small sample sizes (less than 30) and unknown population standard deviation. There are three main types of T-tests:

     One-Sample T-test: Compares the mean of one sample to a known population mean.

      Hence, the margin of error determines the extent of uncertainty in the estimate and directly influences the width of the confidence interval, affecting how precise and reliable the estimate is.

20.  What is an ANOVA test, and what are its assumptions?
   - ANOVA (Analysis of Variance) is a statistical test used to determine if there are significant differences between the means of three or more groups. It helps to assess whether any of the group means differ significantly from each other, based on sample data.

      Assumptions of ANOVA:

      Independence:            
      The samples must be independent of each other. Each observation should be independent of the others.

      Normality:         
      The data within each group should be approximately normally distributed. This can be checked using normality tests or graphical methods like Q-Q plots.

      Homogeneity of Variance:         
      The variances of the populations should be equal across the groups. This can be tested using methods like Levene's test or Bartlett's test.

      When these assumptions are met, ANOVA provides reliable results in determining if there are significant differences between the group means. If any of these assumptions are violated, the results of the ANOVA test may not be valid.



21. What are the different types of ANOVA tests?
   - There are several types of ANOVA (Analysis of Variance) tests, each suited for different scenarios:

     One-Way ANOVA:          
Compares the means of three or more independent groups based on one factor.   
      Example: Comparing the average test scores of students from three different schools.

     Two-Way ANOVA:            
Compares the means of groups based on two factors, allowing for the examination of interaction effects between the factors.     
Example: Studying the effect of teaching methods and gender on student performance.

     Repeated Measures ANOVA:           
Compares the means of the same group measured at different times or under different conditions.            
Example: Assessing the impact of a drug on patients' health at multiple time points.

        Mixed-Design ANOVA (or Split-Plot ANOVA):         
Combines features of both independent and repeated measures ANOVA, examining the effects of between-subjects and within-subjects factors.     
Example: Evaluating the effect of a training program on employees' performance over time, considering different departments.

     Each type of ANOVA is used to analyze variance in different contexts, helping researchers understand the factors influencing their data and identify significant differences between groups.

22. What is the F-test, and how does it relate to hypothesis testing?
   - The F-test is a statistical method used to compare the variances of two or more groups to see if they are significantly different. It is based on the F-distribution and is commonly used in analysis of variance (ANOVA) and regression analysis.

      In hypothesis testing, the F-test starts by formulating two hypotheses. The null hypothesis (Hâ‚€) states that the variances of the groups are equal, while the alternative hypothesis (Hâ‚) states that the variances are not equal.

     To conduct the test, you calculate the F-statistic, which is the ratio of the variances of the groups being compared. Next, you determine the degrees of freedom for both the numerator and the denominator. Using the F-distribution table, you find the critical value based on the significance level (Î±) and the degrees of freedom. Finally, you compare the F-statistic to the critical value. If the F-statistic is greater than the critical value, you reject the null hypothesis, indicating that the variances are significantly different.

     Hence, the F-test helps determine if there are significant differences in variances between groups, making it a valuable tool in hypothesis testing for ANOVA and regression analysis.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
# Ans

import numpy as np
from scipy import stats
sample_data = [5, 7, 8, 6, 9, 7, 10, 6, 8, 9]
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
sample_size = len(sample_data)
population_mean = 7
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
alpha = 0.05
if p_value < alpha:
    result = "Reject the null hypothesis. The sample mean is significantly different from the population mean."
else:
    result = "Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean."
print(f"Sample Mean: {sample_mean}")
print(f"Population Mean: {population_mean}")
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")
print(f"Result: {result}")
```

4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
# Ans

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
np.random.seed(42)
sample_data = np.random.normal(loc=10, scale=2, size=50)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
sample_size = len(sample_data)
population_mean = 10
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha/2)
if p_value < alpha:
    result = "Reject the null hypothesis. The sample mean is significantly different from the population mean."
else:
    result = "Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean."
print(f"Sample Mean: {sample_mean}")
print(f"Population Mean: {population_mean}")
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")
print(f"Result: {result}")
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)
plt.plot(x, y, label="Normal Distribution")
plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.5, label="Rejection Region")
plt.fill_between(x, y, where=(x > -critical_value) & (x < critical_value), color='green', alpha=0.5, label="Non-Rejection Region")
plt.axvline(z_score, color='blue', linestyle='--', label="Z-score")
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.title('Two-Tailed Z-test Decision Regions')
plt.legend()
plt.show()
```

5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
# Ans

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
def visualize_errors(mu0, mu1, sigma, alpha, sample_size):
    z_alpha = stats.norm.ppf(1 - alpha)
    z_beta = (mu1 - mu0) / (sigma / np.sqrt(sample_size)) - z_alpha
    beta = stats.norm.cdf(z_beta)
    power = 1 - beta
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    y0 = stats.norm.pdf(x, loc=mu0, scale=sigma)
    y1 = stats.norm.pdf(x, loc=mu1, scale=sigma)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label=f'Distribution under H0 (Âµ={mu0})', color='blue')
    plt.plot(x, y1, label=f'Distribution under H1 (Âµ={mu1})', color='green')
    plt.fill_between(x, y0, where=(x >= mu0 + z_alpha * (sigma / np.sqrt(sample_size))), color='red', alpha=0.5, label='Type 1 Error (Î±)')
    plt.fill_between(x, y1, where=(x < mu0 + z_alpha * (sigma / np.sqrt(sample_size))), color='orange', alpha=0.5, label='Type 2 Error (Î²)')
    plt.axvline(mu0 + z_alpha * (sigma / np.sqrt(sample_size)), color='black', linestyle='--', label='Critical Value')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.legend()
    plt.show()
    print(f"Type 1 Error (Î±): {alpha}")
    print(f"Type 2 Error (Î²): {beta}")
    print(f"Power of the Test (1-Î²): {power}")
visualize_errors(mu0=0, mu1=2, sigma=1, alpha=0.05, sample_size=30)
```

7.  Perform a paired sample T-test using Python and visualize the comparison results.

9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
# Ans

import numpy as np
from scipy import stats
def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)
    standard_error = sample_std / np.sqrt(sample_size)
    critical_value = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)
    margin_of_error = critical_value * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return (lower_bound, upper_bound)
data = [22, 24, 19, 21, 23, 20, 25, 22, 24, 21]
ci = confidence_interval(data, confidence=0.95)
print(f"95% Confidence Interval: {ci}")

# Significance:
# A confidence interval provides a range of values that is likely to contain the true population parameter
# (e.g., mean) with a specified level of confidence (e.g., 95%). It gives an estimate of the precision of the
# sample mean and indicates how reliable the sample mean is as an estimate of the population mean.
#  A narrower confidence interval suggests greater precision, while a wider interval indicates less precision.

# Hence, confidence intervals help researchers and decision-makers assess the reliability of their estimates and make
# informed conclusions about the population based on sample data.

```

```python
# Ans

prior_prob_disease = 0.01
prob_pos_given_disease = 0.99
prob_pos_given_no_disease = 0.05
prob_no_disease = 1 - prior_prob_disease
prob_pos = (prob_pos_given_disease * prior_prob_disease) + (prob_pos_given_no_disease * prob_no_disease)
posterior_prob_disease_given_pos = (prob_pos_given_disease * prior_prob_disease) / prob_pos
print(f"Prior Probability of Disease (P(Disease)): {prior_prob_disease}")
print(f"Probability of Positive Test given Disease (P(Pos|Disease)): {prob_pos_given_disease}")
print(f"Probability of Positive Test given No Disease (P(Pos|NoDisease)): {prob_pos_given_no_disease}")
print(f"Marginal Probability of Positive Test (P(Pos)): {prob_pos}")
print(f"Posterior Probability of Disease given Positive Test (P(Disease|Pos)): {posterior_prob_disease_given_pos}")
```

   Calculate Marginal Probability (P(B)): Calculate the total probability of a positive test considering both the presence and absence of the disease.

   Calculate Posterior Probability (P(A|B)): Use Bayes' Theorem to update the probability of having the disease given a positive test result.

   Print the Result: Output the prior probability, likelihoods, marginal probability, and posterior probability.

   This process demonstrates how Bayesian inference updates the probability of a hypothesis (having the disease) based on new evidence (positive test result), providing a mathematical framework for combining prior knowledge with new data to make informed decisions.

13.  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.

```python
# Ans

import numpy as np
def calculate_expected_frequencies(observed_data):
    observed_data = np.array(observed_data)
    row_totals = observed_data.sum(axis=1, keepdims=True)
    column_totals = observed_data.sum(axis=0, keepdims=True)
    total = observed_data.sum()
    expected_frequencies = (row_totals @ column_totals) / total
    return expected_frequencies
observed_data = [
    [30, 10],
    [20, 40]
]
expected_frequencies = calculate_expected_frequencies(observed_data)
print("Expected Frequencies:")
print(expected_frequencies)
```

```python
# Ans

import numpy as np
from scipy import stats
data = [22, 24, 19, 21, 23, 20, 25, 22, 24, 21]
hypothesized_variance = 4.0
sample_variance = np.var(data, ddof=1)
n = len(data)
chi_square_statistic = (n - 1) * sample_variance / hypothesized_variance
df = n - 1
p_value = 2 * (1 - stats.chi2.cdf(chi_square_statistic, df))
alpha = 0.05
chi_square_critical = stats.chi2.ppf(1 - alpha/2, df)
if chi_square_statistic > chi_square_critical or chi_square_statistic < stats.chi2.ppf(alpha/2, df):
    result = "Reject the null hypothesis. The population variance is significantly different from the hypothesized variance."
else:
    result = "Fail to reject the null hypothesis. The population variance is not significantly different from the hypothesized variance."
print(f"Sample Variance: {sample_variance}")
print(f"Hypothesized Variance: {hypothesized_variance}")
print(f"Chi-square Statistic: {chi_square_statistic}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value}")
print(f"Result: {result}")
```
